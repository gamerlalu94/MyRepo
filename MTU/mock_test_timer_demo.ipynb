# Mock Test Question Timer - Setup & Demo

This notebook provides a comprehensive guide to setting up and using the Mock Test Question Timer application.

## Quick Start
The application tracks time spent on different types of questions during mock tests and logs all data to CSV files for analysis.

## 1. Installation & Requirements

### Dependencies
The application uses only Python standard libraries:
- `tkinter` - GUI framework (built-in with Python)
- `threading` - Background timer
- `csv` - Data logging
- `datetime` - Timestamps
- `pathlib` - File handling
- `dataclasses` - Data models (Python 3.7+)

### Installation
1. Ensure Python 3.8+ is installed
2. Clone or download `mock_test_timer_gui.py`
3. No external package installation needed!

## 2. Running the Application

### From Command Line
```bash
python mock_test_timer_gui.py
```

### From Python
```python
from mock_test_timer_gui import main
main()
```

### Key Points
- Application opens as a popup window
- Runs in background while taking mock test
- Does NOT block other applications
- Can be minimized/restored as needed

## 3. Initial Configuration Screen

### Screen 1: Setup Options

When you first launch the application, you'll see three configuration options:

- **Verification** (2-30 min): Time to verify and double-check

**Tips**:
- Set based on your target exam pace
- Start with 2 minutes per category
- Adjust based on difficulty

### Option 3: GUI Theme
**Purpose**: Choose visual appearance

**Available Themes**:
1. **Light** - White background (default, good for daytime)
2. **Dark** - Dark background (reduces eye strain)
3. **Professional** - Blue professional look

````python
# Example: Reading and analyzing the CSV file
import pandas as pd
import os
from pathlib import Path

# Find the latest CSV file
log_dir = Path("mock_test_logs")
if log_dir.exists():
    csv_files = list(log_dir.glob("*.csv"))
    if csv_files:
        latest_file = max(csv_files, key=os.path.getctime)
        print(f"Latest log file: {latest_file}")
        
        # Read the CSV
        df = pd.read_csv(latest_file, skiprows=2)  # Skip header rows
        print("\nData Summary:")
        print(df.head())
    else:
        print("No CSV files found yet. Run the application first!")
else:
    print("mock_test_logs directory not found. Run the application first!")
````

## 6. Data Analysis After Test

### Key Metrics to Calculate

### 1. Time Efficiency
```python
# Questions over allocated time
df['Over Time'] = df['Actual (min)'] - df['Allocated (min)']
df['Efficiency %'] = (df['Allocated (min)'] / df['Actual (min)']) * 100
```

### 2. Category Analysis
```python
# Average time by category
category_avg = df.groupby('Time Category')['Actual (min)'].mean()
```

### 3. Theme Analysis
```python
# Average time by theme/subject
theme_avg = df.groupby('Theme')['Actual (min)'].mean()
```

### 4. Performance Stats
```python
total_time = df['Actual (min)'].sum()
avg_per_question = df['Actual (min)'].mean()
longest_question = df['Actual (min)'].max()
quickest_question = df['Actual (min)'].min()
```

````python
# Example Analysis Code
import pandas as pd
import matplotlib.pyplot as plt

# Sample data (replace with actual CSV)
data = {
    'Question #': [1, 2, 3, 4, 5],
    'Time Category': ['Thinking', 'Solving', 'Applying', 'Verification', 'Thinking'],
    'Allocated (min)': [2, 2, 2, 2, 2],
    'Actual (min)': [2.5, 3.2, 1.8, 0.9, 4.1],
    'Theme': ['Physics', 'Physics', 'Physics', 'Chemistry', 'Physics']
}

df = pd.DataFrame(data)

print("=" * 50)
print("MOCK TEST ANALYSIS REPORT")
print("=" * 50)

print(f"\nTotal Questions: {len(df)}")
print(f"Total Time Spent: {df['Actual (min)'].sum():.2f} minutes")
print(f"Average Time/Question: {df['Actual (min)'].mean():.2f} minutes")
print(f"Fastest Question: {df['Actual (min)'].min():.2f} minutes")
print(f"Slowest Question: {df['Actual (min)'].max():.2f} minutes")

print("\nTIME BY CATEGORY:")
category_stats = df.groupby('Time Category')['Actual (min)'].agg(['mean', 'count', 'sum'])
print(category_stats)

print("\nTIME COMPARISON (Actual vs Allocated):")
df['Difference'] = df['Actual (min)'] - df['Allocated (min)']
print(df[['Question #', 'Allocated (min)', 'Actual (min)', 'Difference']])
````

## 8. Troubleshooting

### Problem: Application won't start
**Solution**: 
- Ensure Python 3.8+ is installed
- Check tkinter is available: `python -m tkinter`
- Run from command line for error messages

### Problem: Timer not working
**Solution**:
- Ensure threading module is available
- Check system resources
- Restart application

### Problem: CSV file not created
**Solution**:
- Check write permissions in directory
- Ensure `mock_test_logs/` folder can be created
- Run as administrator if needed

### Problem: Data not saving
**Solution**:
- Check CSV file is not open in Excel
- Ensure you click "Stop & Save" not just "Reset"
- Check file permissions

### Problem: GUI looks distorted
**Solution**:
- Try different theme
- Restart application
- Check screen DPI settings
